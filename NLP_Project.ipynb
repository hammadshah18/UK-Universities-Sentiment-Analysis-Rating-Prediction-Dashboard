**converting slang into a meaning**


```python
#convert txt file into dict
data_dict = {}

with open('slang.txt', 'r') as file:
    for line in file:
        line = line.strip()
        if not line:
            continue
        key, value = line.split('=', 1)  # split only at first '='
        data_dict[key] = value.strip()

print(data_dict)
```

```python
chat_word=data_dict
chat_word
```

```python
#converting slang into a normal form
def chat_conversation(text):
  new_text=[]
  for w in text.split():
    if w.upper() in chat_word:
      new_text.append(chat_word[w.upper()])
    else:
      new_text.append(w)
  return ' '.join(new_text)
```

```python
uni['Review']=uni['Review'].apply(chat_conversation)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from scipy.sparse import hstack
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Filter out rows where the 'Review' is an empty string
rating_filtered = rating[rating["Review"].str.strip() != ""]

# Split before fitting vectorizers/encoders (important to avoid data leakage)
X_train, X_test, y_train, y_test = train_test_split(
    rating_filtered[["Review", "University", "City"]],
    rating_filtered["Rating"],
    test_size=0.2,
    random_state=42
)

# Vectorize text
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,3)) # Increased max_features
X_train_text = vectorizer.fit_transform(X_train["Review"])
X_test_text  = vectorizer.transform(X_test["Review"])

# Encode categorical
encoder = OneHotEncoder(handle_unknown="ignore")
X_train_cats = encoder.fit_transform(X_train[["University", "City"]])
X_test_cats  = encoder.transform(X_test[["University", "City"]])

# Combine
X_train_final = hstack([X_train_text, X_train_cats])
X_test_final  = hstack([X_test_text, X_test_cats])

# Train model
model = XGBRegressor(n_estimators=300, learning_rate=0.1, max_depth=8, random_state=42)
model.fit(X_train_final, y_train)

# Evaluate on test set
y_pred = model.predict(X_test_final)


print("MAE:", mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R²:", r2_score(y_test, y_pred))
```
